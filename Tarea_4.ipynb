```python
%pylab inline
from subprocess import call
import glob
import re
import numpy as np
f1=open('Nombres.dat','w')
Exps1=glob.glob("experiment_theta*")
for i in range(1000):
    temp=Exps1[i]
    f1.write(temp)
    f1.write("\n")
f1.close()

    
    
f2=open('Theta_Phi.dat','w')
Exps=glob.glob("experiment_theta*")
for i in range(1000):
    temp=Exps[i].replace('\n','')
    temp=Exps[i].replace('experiment_theta_','')
    temp=temp.replace('.dat','')
    temp=temp.replace('_phi','')
    temp=temp.replace('_',' ')
    f2.write(temp)
    f2.write("\n")
f2.close()
```

```python
F=[]
suma=0;
gm=0;
for i in range(1000):
    suma=suma+g[i];
gm=suma/1000
for i in range(1000):
    F.append(1-g[i]/gm)
```
